```python
from pydantic import BaseModel, Field
from typing import List, Literal

class Entity(BaseModel):
  entity_value: str = Field(...,description="The actual name or value of the entity.")
  entity_type: Literal["M", "F", "O"] = Field(...,description="The type of the entity, where 'M' stands for male, 'F' for female, and 'O' for other.")
```

```python
data = {
  "entity_value": "John Doe"}
entity = Entity(**data)
print(entity)
```

```python
def func(info:Entity):
  print(info.entity_value)
func({4,"M"})
```
